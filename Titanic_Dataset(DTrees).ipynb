```python
import pandas as pd
df=pd.read_csv('Downloads/Titanic-Dataset.csv')
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.sample(n=10)
```

```python
X=df[['Pclass','Sex','Age','Fare']]
X
```

```python
y=df['Survived']
y
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X['Sex_n']=le.fit_transform(X['Sex'])
X
```

```python
X_n=X.drop(['Sex'],axis='columns')
X_n
```

```python
len(X_n)
```

```python
len(X_n)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_n,y,test_size=0.2,random_state=42)
```

```python
len(X_train)
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict(X_test)
```

```python
model.predict([[1,25.0,30.0,1]])
```

```python
y_predicted = model.predict(X_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predicted)
cm
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
